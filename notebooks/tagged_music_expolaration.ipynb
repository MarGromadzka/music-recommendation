```python
joined = tagged_music.set_index('id').join(songs_len.set_index('id'), on='id', lsuffix='_caller', rsuffix='_other')
```

```python
joined
```

```python
TAGS_LIST = ['guitar', 'classical', 'slow', 'techno', 'strings', 'drums', 'electronic', 'rock', 'fast', 'piano', 'ambient', 'beat', 'violin', 'vocal', 'synth', 'female', 'indian', 'opera', 'male', 'singing', 'vocals', 'no vocals', 'harpsichord', 'loud', 'quiet', 'flute', 'woman', 'male vocal', 'no vocal', 'pop', 'soft', 'sitar', 'solo', 'man', 'classic', 'choir', 'voice', 'new age', 'dance', 'male voice', 'female vocal', 'beats', 'harp', 'cello', 'no voice', 'weird', 'country', 'metal', 'female voice', 'choral']
```

```python
for tag in TAGS_LIST:
  joined[tag] = joined[tag] * joined['no_spec']
```

```python
joined
```

```python
sum = {tag : joined[tag].sum() for tag in TAGS_LIST}
```

```python
sum
```

```python
from operator import itemgetter
sorted(sum.items(), key=itemgetter(1), reverse=True)
```

```python
tags_to_del = []
for key, item in sum.items():
  if item < 16000:
    tags_to_del.append(key)
```

```python
tags_to_del
```

```python
common_tags = tagged_music.drop(columns=tags_to_del)
```

```python
common_tags.to_csv('/content/gdrive/MyDrive/musicrecom/melodice/tagged_music_after_drop')
```

```python
for tag in tags_to_del:
  del sum[tag]
```
